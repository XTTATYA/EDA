```python
def brackets(df,column):
  df[column] = df[column].str.replace(r'\[.*?\]', '', regex=True).str.strip()
  return df[column]
df['name']=brackets(df,'name')
```

```python
Q1 = df['selling_price'].quantile(0.25)
Q3 = df['selling_price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['selling_price'] < lower_bound) | (df['selling_price'] > upper_bound)]
print(outliers)
#there are some outliers present in selling price
```

```python
sns.boxplot(df['selling_price'])
plt.title("Distribution of Bike Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

median_price = df['selling_price'].median()

# Step 3: Replace outliers with median
df['selling_price'] = df['selling_price'].apply(lambda x: median_price if x < lower_bound or x > upper_bound else x)
```

```python
#Q11) Identify and remove outliers in the km_driven column using the IQR method?
"""Ans"""
sns.boxplot(data=df['km_driven'])

q1=df['km_driven'].quantile(0.25)
q3=df['km_driven'].quantile(0.75)
IQR=q3-q1
lower=q1-1.5*IQR
upper=q3+1.5*IQR
outliers=[km for km in df['km_driven'] if km < lower or km > upper]
outliers


median= df['km_driven'].median()

df.loc[df['km_driven'].isin(outliers), 'km_driven'] = median
```
