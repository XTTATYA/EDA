```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv('CarSale.csv')
df
```

```python
df.drop(['Car_id'],axis=1,inplace=True)
```

```python
df['Customer_Name']=df['Customer Name']
df['Annual_Income']=df['Annual Income']
df['Price']=df['Price ($)']
df['Body_Style']=df['Body Style']
```

```python
df.info()
```

```python
df.drop(['Customer Name','Annual Income','Price ($)','Body Style'],axis=1,inplace=True)
```

```python
df
```

```python
#customer name has one null value
df.loc[df['Customer_Name'].isna()]
df['Customer_Name'].fillna('Not provided',inplace=True)
```

```python
df.info()
```

```python
#now for checking for duplicates
df.duplicated().sum()
```

```python
#now lets convert the columns based on appropriate datatype
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df.drop(['Date'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.Engine.unique()
dic={
    'DoubleÂ\xa0Overhead Camshaft':'DOHC',
    'Overhead Camshaft':'OHC'
}
df['Engine']=df['Engine'].map(dic)
```

```python
#checking for outliers in price adn annual income
sns.boxplot(data=df['Annual_Income'])#has n no.of outliers
median=df["Annual_Income"].median()
Q1=df["Annual_Income"].quantile(0.25)
Q3=df["Annual_Income"].quantile(0.75)
IQR=Q3-Q1
lower_bond=Q1-1.5*IQR
upper_bond=Q3+1.5*IQR
df['Annual_Income']=df['Annual_Income'].apply(lambda x:median if x>upper_bond or x<lower_bond else x)
```

```python
sns.boxplot(data=df['Price'])
```

```python
median=df["Price"].median()
Q1=df['Price'].quantile(0.25)
Q3=df['Price'].quantile(0.75)
IQR=Q3-Q1
lower_bond=Q1-1.5*IQR
upper_bond=Q3+1.5*IQR
df['Price']=df['Price'].apply(lambda x:median if x>upper_bond or x<lower_bond else x)
```

```python
df.info()
```

```python
#Q1) What is the average selling price of cars for each dealer, and how does it compare across different dealers?
"""Ans"""
avg_price=df.groupby('Dealer_Name')['Price'].mean()
avg_price
```

```python
plt.figure(figsize=(15,15))
sns.barplot(x='Dealer_Name',y="Price",data=df,hue='Dealer_Name',estimator='mean',palette='coolwarm',ci=None)
plt.title("Average Price Based On Dealers")
plt.xticks(rotation=90)
plt.ylabel("Average Price")
plt.xlabel=("Dealer Names")
plt.show()
#insights are that the average selling price for a car based on each dealer name is around 25000 dollars
```

```python
#Q2)Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends?
"""Ans"""
variation = df.groupby('Company')['Price'].std().reset_index().sort_values(by='Price', ascending=False)

print(variation.head(1))


# Line plot
plt.figure(figsize=(15,10))
sns.lineplot(x='Company', y='Price', data=variation, marker='o',palette='Set3')
plt.xticks(rotation=60)
plt.ylabel('Price Standard Deviation')
plt.title('Price Variation Across Car Brands')
plt.tight_layout()
plt.show()


#insights are that pontiac highly variates in price while hyundai has the least variation
```

```python
#Q3)What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?
"""Ans"""
sns.boxplot(x='Transmission', y='Price', data=df, palette='Set2')
plt.title("Car Price Distribution by Transmission Type")
plt.ylabel("Price")

plt.show()
```

```python
iqr_values = df.groupby('Transmission')['Price'].quantile([0.25, 0.75]).unstack()
iqr_values['IQR'] = iqr_values[0.75] - iqr_values[0.25]
print(iqr_values[['IQR']])
```

```python
#Q4)What is the distribution of car prices across different regions?
"""Ans"""
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Price',hue='Dealer_Region',kde=True,palette='coolwarm')
plt.title("Distribution of Prices based on Regions")
plt.ylabel('Dealer Regin')

plt.show()
#insights are that it is a right skewed distribution.
```

```python
#Q5)What is the distribution of cars based on body styles?
"""Ans"""

sns.countplot(x='Body_Style', data=df, order=df['Body_Style'].value_counts().index, palette='Set3')
plt.title("Distribution of Cars by Body Style")

plt.ylabel("Number of Cars")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#insights are that most companies have the suv type of body style for their cars
```

```python
#Q6)How does the average selling price of cars vary by customer gender and annual income
avg_price=df.groupby(["Gender",'Annual_Income'])['Price'].mean().reset_index()
print(avg_price)
plt.figure(figsize=(10,10))

sns.barplot(x=pd.cut(df['Annual_Income'], bins=[0, 300000, 600000, 900000, 1200000, np.inf], #here pd.cut is used to cut continuos numeric valu into discrete values and np.inf ensures that larger number above finite value are coverd
             labels=['<3L', '3-6L', '6-9L', '9-12L', '12L+']), y='Price', hue='Gender',data=avg_price, palette='Set3',ci=None)
plt.xticks(rotation=90)
plt.title('Average Car Price by Gender and Income')
plt.ylabel('Average Price')
plt.show()
```
