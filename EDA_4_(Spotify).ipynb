```python
#Q2) What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram?
"""Ans"""
plt.figure(figsize=(8,6))
sns.histplot(x='Popularity',data=df,kde=True,palette='coolwarm')
plt.title('Distribution of popularity in dataset')
plt.ylabel('Count of tracks')
plt.show()
#it is kind of a left skewed distribution
```

```python
#Q3) Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot?
"""Ans"""
plt.figure(figsize=(8,6))
sns.scatterplot(x='Popularity',y='Duration (ms)',data=df)
plt.title('Distribution of popularity in dataset')
plt.ylabel('Duration')
plt.show()
#if the duration of song is more than 4 min, the sone loses its popularity
```

```python
#Q4) Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot?
"""Ans"""
plt.figure(figsize=(20,10))
sns.countplot(x='Artist',data=df,palette='coolwarm',)
plt.title('Highest no.of tracks by an Artist')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q5)What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each?
"""Ans"""
v=df[['Popularity','Artist','Track Name']].sort_values(by='Popularity',ascending=True)
v.head(5)
```

```python
#Q6)Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist?
"""Ans"""
plt.figure(figsize=(20,10))
sns.barplot(x='Artist',y='Popularity',data=df,palette='pastel',ci=None,estimator='mean')
plt.title('Popularity of each Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
n=df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)
n
```

```python
#Q7)For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist?
"""Ans"""
n.head(5)
top_5_artists = n.head(5)
top_5_artist_names = top_5_artists.index

filtered_df = df[df['Artist'].isin(top_5_artist_names)]

most_popular_tracks = filtered_df.sort_values(['Artist', 'Popularity'], ascending=[True, False]) \
                                 .drop_duplicates(subset=['Artist']) \
                                 [['Artist', 'Track Name', 'Popularity']]

print(most_popular_tracks)
```

```python
#Q8)Visualize relationships between multiple numerical variables simultaneously using a pair plot?
"""Ans"""
num_cols = df.select_dtypes(exclude='object')
num_cols
plt.figure(figsize=(20,10))
sns.pairplot(data=num_cols,palette='pastel')
```

```python
#Q9)Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot?
"""Ans"""
plt.figure(figsize=(20,10))
sns.boxplot(data=df,x='Artist',y='Duration (ms)',palette='pastel')
plt.title("Artist Duration")
plt.xticks(rotation=90)
plt.show()
#no,most artist have around the same duration of around 2-4 mins
```

```python
#Q10)How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot?
"""Ans"""

plt.figure(figsize=(20, 10))
sns.violinplot(x='Artist', y='Popularity', data=df, palette='Set3')
plt.title('Distribution of Track Popularity by Artist')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
#the more popular the artist the more popular the track
```
